```python
import numpy as np
import matplotlib.pyplot as plt

random_numbers = np.random.uniform(low=0.0, high=1.0, size=1000)

plt.hist(random_numbers, bins=100, edgecolor='black')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Uniform Random Numbers (0,1)")

plt.savefig("uniform_histogram.pdf")

plt.show()
```
