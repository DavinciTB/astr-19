```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)

def exp_func(val):
    return np.exp(val)

y = exp_func(x)

plt.figure(figsize=(8, 6)) # Optional: set figure size for better visibility
plt.plot(x, y, label='exp(x)')

plt.xlabel("Time [milliseconds]", fontsize=12)
plt.ylabel("Awesomeness", fontsize=12)

plt.title("Awesomeness Over Time", fontsize=14)
plt.legend()

plt.savefig("Awesomeness_Plot.pdf")

plt.show()
```
